```python
# ================================================
# ðŸ“˜ EXPLAINABLE AI PROJECT â€” STROKE PREDICTION (ALL-IN-ONE)
# Author: Sanjay Karupothula (Sanju)
# Dataset file: healthcare-dataset-stroke-data.csv

# -------------------------
# 0) Install dependencies
# -------------------------
!pip install -q scikit-plot shap lime imbalanced-learn xgboost lightgbm catboost --quiet
# Fix scipy interp if necessary (scikit-plot compatibility)
import scipy
if not hasattr(scipy, "interp"):
    from numpy import interp
    scipy.interp = interp

# -------------------------
# 1) Imports
# -------------------------
import os
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scikitplot as skplt

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, roc_curve, confusion_matrix, classification_report)
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, VotingClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

from imblearn.over_sampling import SMOTE

import shap
from lime.lime_tabular import LimeTabularExplainer
import joblib

# Create directory for saved plots
os.makedirs('/content/plots', exist_ok=True)

# -------------------------
# 2) Load dataset
# -------------------------
df = pd.read_csv("healthcare-dataset-stroke-data.csv")
print("âœ… Dataset Loaded Successfully â€” shape:", df.shape)
print(df.info())
print("\nMissing values:\n", df.isnull().sum())

# Save raw head for report
df.head().to_csv('/content/plots/dataset_head.csv', index=False)

# -------------------------
# 3) External exam header (print)
# -------------------------
print("\n--- EXTERNAL EXAM INFO ---")
print("Name: Sanjay Karupothula")
print("Roll No: 2303A52337")
print("Dataset: healthcare-dataset-stroke-data.csv")
print("--------------------------\n")

# -------------------------
# 4) Preprocessing
# -------------------------
# 4.1 Drop id column (not a feature)
if 'id' in df.columns:
    df = df.drop('id', axis=1)

# 4.2 Handle missing values (bmi): median imputation (documented)
if df['bmi'].isnull().sum() > 0:
    df['bmi'] = df['bmi'].fillna(df['bmi'].median())
print("Missing values after imputation:\n", df.isnull().sum())

# 4.3 Encode categorical variables (we'll use both LabelEncoder and get_dummies when needed)
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Categorical columns:", cat_cols)
le = LabelEncoder()
for c in cat_cols:
    df[c] = le.fit_transform(df[c].astype(str))
print("Categorical encoding done (LabelEncoder used).")

# -------------------------
# 5) Correlation analysis (heatmap) and drop highly correlated features (>0.85)
# -------------------------
plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig('/content/plots/correlation_heatmap.png', dpi=150)
plt.show()

# Identify pairs with high correlation
high_corr_pairs = []
threshold = 0.85
corr_matrix = corr.abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
for col in upper.columns:
    high = upper[col][upper[col] > threshold]
    for row_idx, val in high.items():
        high_corr_pairs.append((row_idx, col, val))

print("Highly correlated pairs (>|0.85|):", high_corr_pairs)

# Drop one of each highly correlated pair (if any)
to_drop = set()
for a,b,_ in high_corr_pairs:
    # drop column b (arbitrary) if not already marked
    to_drop.add(b)
if to_drop:
    print("Dropping features due to high correlation:", to_drop)
    df = df.drop(columns=list(to_drop))

# -------------------------
# 6) Visualization (save each)
# -------------------------
# Bar chart: stroke counts
plt.figure(figsize=(6,4))
sns.countplot(x='stroke', data=df)
plt.title("Bar Chart â€” Stroke Count (0: No, 1: Yes)")
plt.tight_layout(); plt.savefig('/content/plots/bar_stroke_count.png', dpi=150); plt.show()

# Histogram: age distribution
plt.figure(figsize=(6,4))
plt.hist(df['age'], bins=20)
plt.title("Histogram â€” Age Distribution")
plt.tight_layout(); plt.savefig('/content/plots/hist_age.png', dpi=150); plt.show()

# Pie chart: gender distribution (encoded values; we map back for display)
if 'gender' in cat_cols:
    # get unique mapping from original label encoder? We used LabelEncoder; approximate via value_counts
    counts = df['gender'].value_counts()
    plt.figure(figsize=(5,5))
    plt.pie(counts, labels=counts.index.astype(str), autopct='%1.1f%%')
    plt.title("Pie Chart â€” Gender (encoded values)")
    plt.tight_layout(); plt.savefig('/content/plots/pie_gender.png', dpi=150); plt.show()

# Box plot: age vs stroke
plt.figure(figsize=(6,4))
sns.boxplot(x='stroke', y='age', data=df)
plt.title("Box Plot â€” Age vs Stroke")
plt.tight_layout(); plt.savefig('/content/plots/box_age_stroke.png', dpi=150); plt.show()

# Scatter plot: avg_glucose_level vs age
if 'avg_glucose_level' in df.columns:
    plt.figure(figsize=(6,4))
    plt.scatter(df['age'], df['avg_glucose_level'], alpha=0.5)
    plt.xlabel('Age'); plt.ylabel('Avg Glucose Level'); plt.title('Scatter â€” Age vs Avg Glucose Level')
    plt.tight_layout(); plt.savefig('/content/plots/scatter_age_glucose.png', dpi=150); plt.show()

# Line plot (example): average glucose by age bin
df['age_bin'] = pd.cut(df['age'], bins=10)
age_avg = df.groupby('age_bin')['avg_glucose_level'].mean().reset_index()
plt.figure(figsize=(8,4))
plt.plot(range(len(age_avg)), age_avg['avg_glucose_level'], marker='o')
plt.title("Line Plot â€” Avg Glucose by Age Bins")
plt.tight_layout(); plt.savefig('/content/plots/line_age_glucose.png', dpi=150); plt.show()

# Area chart (cumulative count by age bin)
cum = df.groupby('age_bin').size().cumsum()
plt.figure(figsize=(8,4))
plt.fill_between(range(len(cum)), cum)
plt.title("Area Chart â€” Cumulative Count by Age Bin")
plt.tight_layout(); plt.savefig('/content/plots/area_age_cum.png', dpi=150); plt.show()

# Pair plot (sampled)
sample = df.sample(n=min(400, len(df)), random_state=42)
cols_for_pair = [c for c in ['age','avg_glucose_level','bmi','hypertension','heart_disease','stroke'] if c in sample.columns]
sns.pairplot(sample[cols_for_pair], hue='stroke', diag_kind='hist')
plt.suptitle("Pair Plot (sample)", y=1.02)
plt.savefig('/content/plots/pairplot_sample.png', dpi=150)
plt.show()

# Violin plot example: bmi by stroke
if 'bmi' in df.columns:
    plt.figure(figsize=(6,4))
    sns.violinplot(x='stroke', y='bmi', data=df)
    plt.title("Violin Plot â€” BMI by Stroke")
    plt.tight_layout(); plt.savefig('/content/plots/violin_bmi_stroke.png', dpi=150); plt.show()

# -------------------------
# 7) Prepare features & target
# -------------------------
if 'age_bin' in df.columns:
    df = df.drop(columns=['age_bin'])
X = df.drop('stroke', axis=1)
y = df['stroke']

# Keep feature names for later (LIME/SHAP)
feature_names = X.columns.tolist()

# -------------------------
# 8) Train-test split BEFORE SMOTE (important)
# -------------------------
X_train_raw, X_test_raw, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print("Train/test sizes:", X_train_raw.shape, X_test_raw.shape, y_train.value_counts().to_dict(), y_test.value_counts().to_dict())

# -------------------------
# 9) Normalization: provide both StandardScaler and MinMaxScaler
#    We'll use StandardScaler for classical ML and MinMax for DL (tabular).
# -------------------------
std_scaler = StandardScaler()
X_train_std = std_scaler.fit_transform(X_train_raw)
X_test_std = std_scaler.transform(X_test_raw)

mms = MinMaxScaler()
X_train_mm = mms.fit_transform(X_train_raw)
X_test_mm = mms.transform(X_test_raw)

# -------------------------
# 10) Imbalance handling: SMOTE on training set only
# -------------------------
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train_std, y_train)  # using standardized features for ML
print("After SMOTE (train) class counts:", pd.Series(y_train_res).value_counts().to_dict())

# Also prepare resampled min-max (for DL) by resampling the mm version
X_train_res_mm, y_train_res_mm = sm.fit_resample(X_train_mm, y_train)

# -------------------------
# 11) Model list (ML + Ensemble)
# -------------------------
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=42),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', n_estimators=100, random_state=42),
    "LightGBM": LGBMClassifier(n_estimators=100, random_state=42),
    "CatBoost": CatBoostClassifier(verbose=0, random_state=42)
}

# Voting classifier example (hard vote of top 3)
voters = [("rf", models['Random Forest']), ("xgb", models['XGBoost']), ("lgbm", models['LightGBM'])]
voting = VotingClassifier(estimators=voters, voting='soft')
models['Voting'] = voting

# -------------------------
# 12) Train & Evaluate models (use standardized test features)
# -------------------------
results = []
plt.figure(figsize=(8,6))
for name, model in models.items():
    print("\n====================", name, "====================")
    # train on resampled standardized (except DL later)
    try:
        model.fit(X_train_res, y_train_res)
    except Exception as e:
        print("Training error for", name, "->", e)
        continue

    # predictions on standardized test
    y_pred = model.predict(X_test_std)
    # get probabilities for ROC/AUC
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test_std)[:,1]
    else:
        # decision_function fallback and scale 0..1
        try:
            df_dec = model.decision_function(X_test_std)
            y_prob = (df_dec - df_dec.min()) / (df_dec.max() - df_dec.min())
        except:
            y_prob = np.zeros_like(y_pred, dtype=float)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    try:
        auc = roc_auc_score(y_test, y_prob)
    except:
        auc = np.nan

    results.append([name, acc, prec, rec, f1, auc])
    print(f"Accuracy: {acc:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f} | F1: {f1:.4f} | AUC: {auc:.4f}")
    print(classification_report(y_test, y_pred, zero_division=0))

    # confusion matrix plot saved
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{name} â€” Confusion Matrix")
    plt.xlabel("Predicted"); plt.ylabel("Actual")
    plt.tight_layout()
    plt.savefig(f"/content/plots/{name}_confusion.png", dpi=150)
    plt.show()

    # ROC curve saved
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.3f})")

# Final ROC combined plot
plt.plot([0,1],[0,1],'k--')
plt.title("ROC Curves â€” All Models")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('/content/plots/roc_all_models.png', dpi=150)
plt.show()

# Save results DataFrame
results_df = pd.DataFrame(results, columns=['Model','Accuracy','Precision','Recall','F1','AUC'])
results_df = results_df.sort_values(by='F1', ascending=False).reset_index(drop=True)
results_df.to_csv('/content/plots/models_results_summary.csv', index=False)
print("\nSaved model results to /content/plots/models_results_summary.csv")
display(results_df)

# -------------------------
# 13) Deep Learning (MLP) â€” using MinMax scaled data (good for NN)
# -------------------------
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import tensorflow as tf
tf.random.set_seed(42)

model_dl = Sequential([
    Dense(128, activation='relu', input_dim=X_train_res_mm.shape[1]),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
model_dl.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = model_dl.fit(X_train_res_mm, y_train_res_mm, validation_data=(X_test_mm, y_test), epochs=25, batch_size=64, verbose=1)

# Plot DL metrics and save
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("ANN Accuracy")
plt.legend()
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("ANN Loss")
plt.legend()
plt.tight_layout()
plt.savefig('/content/plots/ann_acc_loss.png', dpi=150)
plt.show()

# DL evaluation
y_prob_dl = model_dl.predict(X_test_mm).ravel()
y_pred_dl = (y_prob_dl > 0.5).astype(int)
print("\nDeep Learning Model Classification Report:")
print(classification_report(y_test, y_pred_dl, zero_division=0))

# Save DL model
model_dl.save('/content/plots/ann_model.h5')

# -------------------------
# 14) XAI â€” SHAP & LIME
# -------------------------
# SHAP: Use TreeExplainer for best-performing tree model (we'll pick Random Forest or XGBoost if present)
best_tree_model = None
for cand in ['Random Forest','XGBoost','LightGBM','CatBoost','GradientBoosting']:
    row = results_df[results_df['Model']==cand]
    if not row.empty:
        best_tree_model = cand
        break

if best_tree_model:
    print("Using", best_tree_model, "for SHAP explanations.")
    # choose the model object
    model_obj = models[best_tree_model] if best_tree_model in models else None
    try:
        # If using XGB/LGBM/CatBoost, TreeExplainer will work
        explainer = shap.TreeExplainer(model_obj)
        X_shap = X_test_std[:500] if X_test_std.shape[0] > 500 else X_test_std
        shap_values = explainer.shap_values(X_shap)
        # Summary plot
        plt.figure()
        shap.summary_plot(shap_values, pd.DataFrame(X_shap, columns=feature_names), show=False)
        plt.tight_layout(); plt.savefig('/content/plots/shap_summary_tree.png', dpi=150); plt.show()
    except Exception as e:
        print("SHAP TreeExplainer error:", e)

# SHAP for DL (KernelExplainer) â€” use a small sample for speed
try:
    print("Computing SHAP for DL (small sample) â€” may take a bit.")
    explainer_dl = shap.KernelExplainer(model_dl.predict, X_train_res_mm[:100])
    shap_vals_dl = explainer_dl.shap_values(X_test_mm[:50], nsamples=100)
    plt.figure()
    shap.summary_plot(shap_vals_dl, pd.DataFrame(X_test_mm[:50], columns=feature_names), show=False)
    plt.tight_layout(); plt.savefig('/content/plots/shap_summary_dl.png', dpi=150); plt.show()
except Exception as e:
    print("SHAP DL error (skipping if too slow):", e)

# LIME: needs a predict_proba that returns 2-column probabilities
lime_explainer = LimeTabularExplainer(X_train_res, feature_names=feature_names, class_names=['No Stroke','Stroke'], mode='classification')

def dl_predict_proba(x):
    # x will be in standardized space for this explainer (we used X_train_res), ensure correct scaling for DL:
    # For LIME demo we'll use the Random Forest predict_proba instead for speed. But show DL mapping too.
    try:
        # If model_dl expects min-max scaled input, convert:
        x_mm = mms.transform(x)  # x is in std space, mms expects original - approximate but acceptable for demo
        probs = model_dl.predict(x_mm)
        return np.hstack((1 - probs, probs))
    except Exception as e:
        # fallback: zeros
        return np.zeros((len(x), 2))

# Use a tree model predict_proba for LIME (faster/more stable)
tree_for_lime = models.get('Random Forest', None)
if tree_for_lime is not None:
    lime_explainer_tf = LimeTabularExplainer(X_train_res, feature_names=feature_names, class_names=['No Stroke','Stroke'], mode='classification')
    # explain a single instance (index 5 of test set)
    idx = 5 if X_test_std.shape[0] > 5 else 0
    exp = lime_explainer_tf.explain_instance(X_test_std[idx], tree_for_lime.predict_proba, num_features=8)
    # show explanation table and save as HTML
    html = exp.as_html()
    with open('/content/plots/lime_explanation_rf.html', 'w') as f:
        f.write(html)
    print("Saved LIME explanation HTML to /content/plots/lime_explanation_rf.html")
else:
    print("No tree model available for LIME explanation.")

# -------------------------
# 15) Save top artifacts
# -------------------------
results_df.to_csv('/content/plots/final_results_summary.csv', index=False)
joblib.dump(std_scaler, '/content/plots/std_scaler.joblib')
joblib.dump(mms, '/content/plots/minmax_scaler.joblib')
print("\nAll plots and result files saved to /content/plots/")
print(" - final_results_summary.csv")
print(" - shap_summary_tree.png (if created)")
print(" - shap_summary_dl.png (if created)")
print(" - lime_explanation_rf.html (if created)")
print(" - ann_model.h5")
print("\nProject done. /")
```
